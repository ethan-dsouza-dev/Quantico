```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plot_open(name):
    # Load the CSV-like file (assuming X.ipynb is actually a CSV)
    file_path = f"{name}_data.csv"  # Update this if needed

    data = pd.read_csv(file_path)

    # Plot the Open variable
    plt.figure(figsize=(10, 5))
    plt.plot(data['Open'], label='Open Price', color='blue')
    plt.xlabel('Time')
    plt.ylabel('Price')
    plt.title(f"Open Price for {name} Over Time")
    plt.legend()
    plt.grid()
    plt.show()
```

```python
plot_open("AAPL")
```

```python
plot_open("GOOG")
```

```python
plot_open("NVDA")
```
