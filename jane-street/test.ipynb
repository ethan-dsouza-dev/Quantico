```python
import pandas as pd

# Replace 'your_file.parquet' with the path to your Parquet file
file_path = 'data/train/part-0.parquet'

# Read the Parquet file
df = pd.read_parquet(file_path)

# Display the DataFrame
print(df)
```

```python
df
```

```python
# Select rows where `symbol_id` is 1
filtered_df = df[df['symbol_id'] == 2]
```

```python
filtered_df['time_id']
```

```python
time_ids = filtered_df['time_id'].sort_values()

# Find gaps by checking the difference between consecutive values
gaps = time_ids.diff().loc[time_ids.diff() > 1]

print(f"Gaps in time_id: {gaps}")
```

```python
for i in range(1, 38):
    filtered_df = df[df['symbol_id'] == i]
    time_ids = filtered_df['time_id'].sort_values()

    # Find gaps by checking the difference between consecutive values
    gaps = time_ids.diff().loc[time_ids.diff() > 1]

    print(f"Gaps in time_id: {gaps}")
```
